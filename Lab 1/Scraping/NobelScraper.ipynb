### import necessary modules

```python
import csv
import requests
from bs4 import BeautifulSoup
```

### get the HTML that we are going to parse

```python
url = "https://www.nobelprize.org/prizes/lists/all-nobel-prizes/"
r = requests.get(url)
c = r.content
soup = BeautifulSoup(c, "html.parser")
winners_soup = soup.find_all("div", class_="by_year")
```

### lets me look at the html directly

```python
for winner in winners_soup:
    print(winner.prettify())
    break
```

### seperate the desired data from html tags and each other
In order to turn the desired information into a CSV file, I need to create a double array, where each item in the array is another array containing the awardee(s), field, year, and work. I have **winners_soup** are all the html per three years so the first one is from 2021 to 2019, so I iterate over with a for loop. I find the field and year from the title. An example is: "The Nobel Prize in **Physics 2019**". Next using next_siblings, based off of the last a tag I find separate the awardee(s) names from their work. An example is: Paul R. Milgrom and Robert B. Wilson **|** “for improvements to auction theory and inventions of new auction formats”

```python
# Array that will be fed to CSV file
luareate_csv = [];

# Iterates over div tags with the classes by_year
for winner in winners_soup:

    award_list = winner.h3.a.text.split()
    year = award_list[len(award_list) - 1]

    if 'in' in award_list:
        in_index = award_list.index('in')
        field = ' '.join(award_list[in_index + 1:len(award_list) - 1])
    else:
        field = 'Peace Prize'

    # Finds Year and Field
    for award in winner.find("h3").next_siblings:

        if award.name=="h3":
            award_list = award.text.split()
            year = award_list[len(award_list) - 1]

            if 'in' in award_list:
                in_index = award_list.index('in')
                field = ' '.join(award_list[in_index + 1:len(award_list) - 1])
            else:
                field = 'Peace Prize'

        else:
            # Finds Attendees and Work
            award_text = award.text.split()
            if award_text:

                if award.find('a'):
                    reason = award.find('a').next_sibling
                    reason_list = reason.split()
                    if reason_list[0] == 'and' or reason_list[0] == ',':
                        for word in award.a.find_next_siblings("a"):
                            reason = word.next_sibling.text
                    reason_list = reason.split()
                    break_index = award_text.index(reason_list[0])
                    laureate = ' '.join(award_text[:break_index])
                else:
                    reason = award.text
                    laureate = "Record not found / Not awarded"
                sub_laureate_csv = [laureate, reason, field, year]
                luareate_csv.append(sub_laureate_csv)
break
```

### convert our array into a csv file

```python
with open('nobel_lauretes.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(luareate_csv)
```
